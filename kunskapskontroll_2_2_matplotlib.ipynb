### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# It's possible to plot explicit or implicit.
# Explicit: Let matplot create figures directly. This is the object oriented way.
# Implicit: Do all plotting through PyPlot. This is a faster and simpler.
#
# Normally the Object Oriented approach is the recommended because of the possibilities.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: This is the canvas where all the information is diplayed.
# Axes: The area where data is plotted. A figure can contain multiple axes.
# Axis: X- or y- axis of a plot. 
# Artist: All the objects in the figure, including the figure. Examples on artists are lines, text, asis and labels.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
That depends on the functionality. Most common data types are lists, NumPy arrays and Panda Dataframes.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# first the OOA plot
# create data
x = np.linspace(-4,4,100)
y = x**2

# create figure
fig, ax = plt.subplots()

# plot data
ax.plot(x,y)

# add title and graph info
ax.set_title('y = x^2 (OOA)')
ax.set_xlabel('x')
```

```python
# now the Pyplot approach

# create figure
x = np.linspace(-4, 4, 100) 
y = x**2

# Plot data
plt.plot(x,y)

# add title and graph info
plt.title('y = x^2 (PyPlot)') 
plt.xlabel('x') 
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,10))

# Scatter plot 
ax1.scatter(random_data_x, random_data_y, alpha=0.5) 
ax1.plot(x, y, color='red') 
ax1.set_title('Scatter Plot') 
ax1.set_xlabel('X-axis') 
ax1.set_ylabel('Y-axis')
               
# Bar plot 
ax2.bar(names, values, color=['purple', 'red', 'orange', 'yellow', 'green']) 
ax2.set_title('Bar Plot') 
ax2.set_xlabel('Fruits') 
ax2.set_ylabel('Quantity')

# Show the plots 
plt.tight_layout()
plt.show()
               
```
